* 자료형 정리하기

```python
nums = ['carat', 'depth', 'table', 'x', 'y', 'z'] # 연속형
noms = ['cut', 'color', 'clarity'] # 이산형
y = 'price'
print('연속형 자료 :', nums)
print('범주 / 이산 자료 :', noms)
print('예측할 변수 :', y)
```

* 데이터 정리하기
  * 데이터 소개: 다양한 형태의 다이아몬드 price(가격)을 예측한다. 
  * 자료형 소개
    * 연속 자료형
        * caret: weight of the diamond
        * depth: total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43–79)
        * table: width of top of diamond relative to widest point 
        * x: length in mm 
        * y: width in mm 
        * z: depth in mm 
    * 이산 자료형
        * clarity: a measurement of how clear the diamond is (I1(worst),SI2,SI1,VS2,VS1,VVS2,VVS1,IF(best))
        * cut: quality of the cut (Fair, Good, Very Good, Premium, Ideal)
        * color : diamond colour,
        from J (worst), I, H, G, F, E, to D (best)
    

  * 모델링을 위한 자료형 구분:

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot 은 시간이 오래 걸리기 때문에 일부 샘플을 추출해 그리는게 좋습니다.
sns.pairplot(data = df.sample(100))
```

* `boxplot`으로 범주 속성별 예측값 관계 시각화하기

```python
fig, axes = plt.subplots(1, len(noms), figsize = (12,4))
for col, ax in zip(noms, axes):
    sns.boxplot(data = df, x = col, y = 'price',  ax = ax)
plt.show()
```

* `scatterplot` 이용해 연속 자료형 별 예측값의 관계 시각화하기

```python
fig, axes = plt.subplots(1, len(nums), figsize = (18,4))
for col, ax in zip(nums, axes):
    sns.scatterplot(data = df.sample(500), x = col, y = 'price',  ax = ax)
plt.show()
```

* `hue` 옵션에 `color` 추가해서 그리기 

```python
fig, axes = plt.subplots(1, len(nums), figsize = (18,4))
for col, ax in zip(nums, axes):
    sns.scatterplot(data = df.sample(500), x = col, y = 'price',  ax = ax, hue = 'cut')
plt.show()
```

```python
fig, axes = plt.subplots(1, len(nums), figsize = (18,4))
for col, ax in zip(nums, axes):
    sns.scatterplot(data = df.sample(500), x = col, y = 'price',  ax = ax, hue = 'color')
plt.show()
```

```python
fig, axes = plt.subplots(1, len(nums), figsize = (18,4))
for col, ax in zip(nums, axes):
    sns.scatterplot(data = df.sample(500), x = col, y = 'price',  ax = ax, hue = 'clarity')
plt.show()
```

* `regplot`으로 회귀 시각화

```python
fig, axes = plt.subplots(1, len(nums), figsize = (18,4))
for col, ax in zip(nums, axes):
    sns.regplot(data = df.sample(1000), x = col, y = 'price',  ax = ax, marker = '.')
plt.show()
```

* `residplot`으로 잔차 시각화

```python
fig, axes = plt.subplots(1, len(nums), figsize = (18,4))
for col, ax in zip(nums, axes):
    sns.residplot(data = df.sample(1000), x = col, y = 'price',  ax = ax)
plt.show()
```

* `lmplot`으로 `color`별 회귀 시각화

```python
for col, ax in zip(nums, axes):
    sns.lmplot(data = df.sample(1000), x = col, y = 'price', hue  = 'cut', col = 'cut')
plt.show()
```
