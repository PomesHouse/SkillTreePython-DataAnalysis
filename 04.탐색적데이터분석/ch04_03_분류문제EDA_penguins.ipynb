* 숫자 자료형 외 값 및 개수 확인

```python
for col in ['species', 'island', 'sex']:
    print(col)
    print(df[col].value_counts())  # 해당 속성의 모든 카테고리 반환
```

### **| 변수 요약**

* 모든 속성의 분포 간단히 그리기

```python
fig, axes = plt.subplots(2, 4, figsize = (15, 6))
axes = axes.flatten()
for col, ax in zip(df.columns, axes):
    sns.histplot(data = df, x = col, ax = ax)
plt.show()
```

* 자료형 정리하기

```python
nums = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g'] # 연속형
noms = ['island', 'sex'] # 범주형 
y = 'species'
print('숫자/연속 자료형', nums)
print('범주 자료형', noms)
print('예측할 속성/변수', y)
```

* 데이터 정리하기
  * 데이터 소개: 펭귄의 3가지 종을 구분한다. 
  * 자료형 소개:
    * 연속 자료형
     * 'bill_length_mm' : 부리의 길이 
     * 'bill_depth_mm' : 부리의 너비
     * 'filpper_length_mm' : 날개의 길이
     * 'body_mass_g' : 몸무게
    * 범주 자료형 
     * island : 섬 
     * sex : 성별 
     * species : 종(예측해야 할 값)
  * 모델링을 위한 자료형 구분:



## 데이터 분포
---

### **| 데이터 분포 시각화**

* 연속 자료형 분포 그리기

```python
fig, axes = plt.subplots(1, 4, figsize = (16,4))

for col, ax in zip(nums, axes):
    sns.histplot(data = df, x = col, ax = ax)
plt.show()
```

* 이산 자료형 분포 그리기

```python
fig, axes = plt.subplots(1, len(noms), figsize = (len(noms)*4, 4))

for col, ax in zip(noms, axes):
    sns.histplot(data = df, x = col, ax = ax )
plt.show()
```

### | **기술 통계량 시각화**

* 기술통계량 확인

```python
df.describe()
```

* boxplot 으로 연속 자료형 기술 통계 시각화하기

```python
fig, axes = plt.subplots(1, 4, figsize = (16,4))

for col, ax in zip(nums, axes):
    sns.boxplot(data = df, x = col, ax = ax)
plt.show()
```

* viloinplot 으로 연속 자료형 기술 통계 시각화하기

```python
fig, axes = plt.subplots(1, 4, figsize = (16,4))

for col, ax in zip(nums, axes):
    sns.violinplot(data = df, x = col, ax = ax)
plt.show()
```

## 속성간 관계 파악
---

### **| 상관분석**

* `corr`를 이용해 상관계수 구하기

```python
df.corr()
```

* `style.background_gradient` 이용해 상관계수 시각화 하기

```python
df.corr().style.background_gradient(cmap = 'coolwarm')
# 상관관계가 매우 높은 변수가 있다면 둘 중 하나를 없애고 모델을 돌리는 것도 하나의 방법이다. 
```

* `heatmap` 이용해 상관계수 시각화 하기

```python
sns.heatmap(df.corr(), cmap = 'coolwarm', vmax = 1, vmin = -1, annot = True)
```

* 상관관계가 높은 속성 정리하기




```python
# 정리
# 날개 길이와 몸무게의 양의 상관관계가 상당히 높다. 
```

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot 은 시간이 오래 걸리기 때문에 일부 샘플을 추출해 그리는게 좋습니다.
_ = sns.pairplot(df.sample(100), hue = 'species')
```

* `kdeplot` 이용해 연속 자료형 별 예측값의 분포 시각화 하기

```python
for col in nums:
    sns.displot(data = df, x = col, kind = 'kde', hue = y)
    plt.show()
```

* 위에서 그린 플롯에 `col` 옵션으로 `island` 추가하기

```python
for col in nums:
    sns.displot(data = df, x = col, kind = 'kde', hue = y, col = 'island')
    plt.show()
```

* `bill_length_mm`과 `bill_depth_mm`의 산점도를 그리고 `hue`옵션에 예측값 추가하기

```python
sns.scatterplot(data = df, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species')
```

* `lmplot`을 이용해 위의 산점도를 다시 그리고 `col` 옵션에 `island` 추가하기

```python
sns.lmplot(data = df, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species', col = 'island')
```

* `flipper_length_mm`, `body_mass_g`의 산점도를 그리고 `hue`옵션에 예측값 추가하기

```python
sns.scatterplot(data = df, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'species')
```

* `lmplot`을 이용해 위의 산점도를 다시 그리고 `col` 옵션에 `island` 추가하기

```python
sns.lmplot(data = df, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'species', col = 'island')
```

* `histplot`으로 범주형 속성들과 예측값 히트맵 그리기

```python
# x가 범주/연속, y가 연속(개수)
# 지금은 x 범주, y도 범주(예측값)
fig, axes = plt.subplots(1, len(noms), figsize =  (10, 4))
for col, ax in zip(noms, axes):
    _ = sns.histplot(data = df, x = col, y = y, ax = ax)
```
