* 자료형 정리하기

```python
nums = ['displacement', 'horsepower', 'weight', 'acceleration'] # 연속형 (Continuous)
noms = ['cylinders', 'model_year', 'origin'] # 이산형 (nominal)(숫자의 이상형 vs 범주형) 으로 나눠짐 
y = 'mpg' # 예측값
print('연속형 자료', nums)
print('범주 / 이산 자료', noms)
print('예측할 변수', y)
print('추가 전처리가 필요한 속성 : name' )
```

* 데이터 정리하기
  * 데이터 소개: 1970년 후반 ~ 1980년 초반의 mpg(자동차 연비)를 예측한다. 
  * 자료형 소개
    * 연속 자료형
      * mpg : 자동차 연비 (예측값)
      * displacement: 배기량
      * horspower: 마력
      * acceleration : 가속도
    * 이산 자료형
      * cylinders: 실린더 개수
      * weight: 자동차 무게
      * model_year: 자동차 모델의 연도
    * 범주 자료형
      * origin: 제조국가
      * name: 자동차 브랜드
  * 모델링을 위한 자료형 구분
    * 연속 -> ...
    * 이산 -> ...

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot 시간이 오래걸린다. 따라서 소량의 샘플을 그리는게 좋다. 
df.sample(100) # 데이터 프레임 안에서 샘플 100개 뽑기 
_ = sns.pairplot(df.sample(100), hue = 'origin')
```

* `boxplot`으로 범주 속성별 예측값 관계 시각화하기

```python
# x = 범주 자료형, y = mpg 연비



fig, axes = plt.subplots(1,len(noms), figsize = (12,4))
axes = axes.flatten()
for col, ax in zip(noms, axes):
    sns.boxplot(data = df, x = col, y = 'mpg', ax = ax)
plt.show()
```

* 최신 모델일수록 연비가 좋아진다.

* `scatterplot` 이용해 연속 자료형 별 예측값의 관계 시각화하기

```python
# 연속형과 연속형을 비교
# x = 연속자료형, y = mpg

# noms 리스트 변수 그리기

fig, axes = plt.subplots(1,len(nums), figsize = (12,4))
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.scatterplot(data = df, x = col, y = y, ax = ax)
plt.show()
```

* `hue` 옵션에 `origin` 추가해서 그리기 

```python
fig, axes = plt.subplots(1,len(nums), figsize = (12,4))
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.scatterplot(data = df, x = col, y = y, ax = ax, hue = 'origin')
plt.show()
```

* `regplot`으로 회귀 시각화

```python
# regression 회귀선을 그려준다. 
fig, axes = plt.subplots(1,len(nums), figsize = (12,4))
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.regplot(data = df, x = col, y = y, ax = ax, marker = '.')
plt.show()
```

* `residplot`으로 잔차 시각화

```python
# residplot 잔차
fig, axes = plt.subplots(1,len(nums), figsize = (12,4))
axes = axes.flatten()
for col, ax in zip(nums, axes):
    sns.residplot(data = df, x = col, y = y, ax = ax)
plt.show()
```

* `lmplot`으로 `origin`별 회귀 시각화

```python
for col, ax in zip(nums, axes):
    sns.lmplot(data = df, x = col, y = y, hue = 'origin', col = 'origin')
plt.show()
```
