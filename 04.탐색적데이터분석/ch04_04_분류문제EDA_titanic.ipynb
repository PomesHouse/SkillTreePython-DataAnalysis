### **| 변수 요약**

* 속성을 그리기전 bool 값을 가지는 속성 `adult_mal`과 `alone`을 int 값으로 변환해야 합니다.

```python
df.loc[df['adult_male'] == True,'adult_male'] = 1
df.loc[df['adult_male'] == False,'adult_male'] = 0
df.loc[df['alone'] == True,'alone'] = 1
df.loc[df['alone'] == False,'alone'] = 0
df['adult_male'] = df['adult_male'].astype('int64')
df['alone'] = df['alone'].astype('int64')
```

### **| 속성간 관계 시각화**

* `pairplot` 이용해 시각화 하기 

```python
# pairplot 은 시간이 오래 걸리기 때문에 일부 샘플을 추출해 그리는게 좋습니다.
sns.pairplot(df.sample(100), hue = 'survived')
```

* `displot` 이용해 연속 자료형 별 예측값의 분포 `kde` 선으로 시각화 하기 

```python
for col in nums:
    sns.displot(data = df, x = col, kind = 'kde')
```

* 위에서 그린 플롯에 `col` 옵션으로 `sex` 추가하기

```python
for col in nums:
    sns.displot(data = df, x = col, kind = 'kde', col = 'sex')
```

* 위에서 그린 플롯에 `hue` 옵션으로 `class` 추가하기

```python
for col in nums:
    sns.displot(data = df, x = col, kind = 'kde', col = 'sex', hue = 'class')
```

* `age`과 `fare`의 산점도를 그리고 `hue`옵션에 예측값 추가하기

```python
sns.scatterplot(data = df, x = 'age', y = 'fare', hue = 'survived')
plt.show()
```

* `lmplot`을 이용해 위의 산점도를 다시 그리고 `col` 옵션에 `class` 추가하기

```python
sns.lmplot(data = df, x = 'age', y = 'fare', hue = 'survived', col = 'class')
plt.show()
```

* `histplot`으로 범주형 속성들과 예측값 히트맵 그리기

```python
fig, axes = plt.subplots(3, 4, figsize = (14, 10))
axes = axes.flatten()
for col, ax in zip(noms, axes):
    sns.histplot(data = df, x = col, y = y, ax = ax)
```

```python
 # 범주형 속성들 중 survived와 가장 연관이 있는 속성은 
 # pclass == 3 
 # sex == 'male'
 # sibsp == 0 
 # parch == 0
 # embared == 'S'
 # class == 'Third'
 # who == 'man
 # adult_male == 1
 # embark_town == 'Southampton'
 # alive =='no' (당연)
 # alone == 1
 # 일수록 survived == 0 일 확률이 높을 것이다. 
```
