**예제 01**

다음과 같이 연도, 월, 날짜가 기입된 데이터프레임을 작성하고 `df01` 로 바인딩하세요.

2001년 1-1일 부터 2001년 2-28일까지 기입되어 있습니다. (31+28= 총 59줄)


| index | 연도 | 날짜 | 
|--|--|--|
|0|2001|1-1|
|1|2001|1-2|
|2|2001|1-3|
||...|

````python
years = [2001 for x in range(59)]
date = []
for i in range (1,3):
    for j in range (1,32):
        date.append(f'{i}-{j}')        
        if i == 2 and j == 28:
            break
df01 = pd.DataFrame(columns = ['연도','날짜'])
df01['연도'] = years
df01['날짜'] = date
df01.tail()
````

**예제 02**

`df01`의 연도를 모두 2022로 바꾸세요.

````python
df01['연도'] = 2022
df01.tail()
````

**예제 03**

`df01` 에 1월, 2월에 이어서 3월을 추가하세요.

````python
march = []
for i in range(1,32):
    march.append(['2022'])
    march[i-1].append(f'3-{i}')
march = pd.DataFrame(march, columns = ['연도','날짜'])
march.tail()
````

````python
df01 = pd.concat([df01, march], axis = 0, ignore_index = True)
print(df01.head(10))
df01.tail(10)
````
