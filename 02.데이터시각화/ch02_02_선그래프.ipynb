### **01. 플롯 분할**
---

````python
x = np.linspace(-5, 5, 1000) # -5부터 5까지 1000개 데이터
def func2(a):
    y = a*(x**2)
    return y
# 왼쪽에 
plt.subplot(1, 2, 1) # (행의개수, 열의개수, 위치)
plt.plot(x, func2(1))  # y = x^2 
plt.plot(x, func2(2))

# 오른쪽에
plt.subplot(1, 2 ,2)
plt.plot(x, func2(-1))
plt.plot(x, func2(-2))
# y = -x^2
# y = -2^2
plt.show()
````

### **02. 눈금 공유**
---



````python
t = np.arange(0.01, 5, 0.01) # 0.01부터 5까지 0.01단위 증가 

s1 = np.sin(t *np.pi)
s2 = np.sin(t **np.pi)
s3 = np.exp(-t)# 지수함수

ax1 = plt.subplot(3,1,1)
plt.plot(t, s1)
plt.tick_params('x', labelbottom = False )# 눈금 지우기
plt.subplot(3,1,2, sharex = ax1) # 축공유 
plt.plot(t, s2)
plt.tick_params('x', labelbottom = False )
plt.subplot(3,1,3, sharex = ax1)
plt.plot(t, s3)
````

### **01. 로그함수와 지수함수 그리기**
---

````python
x = np.arange(-5, 5, 0.01)
l = np.log(x)
ml = -(np.log(x))
e = np.exp(x)
me = -(np.exp(x))
plt.subplot(1,2,1)
plt.plot(x, l, 'gray')
plt.plot(x, ml, 'k')
plt.plot(x, np.log2(x), 'darkgray')
plt.plot(x, np.log10(x), 'dimgray')

plt.subplot(1,2,2)
plt.plot(x, e, color = 'mediumpurple')
plt.plot(x, np.exp2(x), 'mediumslateblue')
plt.plot(x, me, color = 'darkslateblue')


# plt.axis([-4,4,-5,5])
````

````python
df = sns.load_dataset('flights')
df.head()
````

````python
plt.plot(df['passengers'])
````

````python
monp = df.groupby('month')['passengers'].mean()
monp
````

````python
plt.plot(monp, color = 'mediumaquamarine', marker = 'o', markersize = 2)
plt.title('Avg. Passengers per Month', fontsize = 15, pad = 10)
plt.xlabel('Month')
plt.ylabel('Number of Avg. Passengers')
plt.tight_layout()
plt.show()
````

````python
yeap = df.groupby('year')['passengers'].sum()
yeap
````

````python
plt.plot(yeap, marker = 'x', markersize = 3, color = 'lightseagreen')
plt.title('Passengers per Year', fontsize = 15, pad = 10)
plt.xlabel('Year')
plt.ylabel('Number of Passenters')
plt.tight_layout()
````
