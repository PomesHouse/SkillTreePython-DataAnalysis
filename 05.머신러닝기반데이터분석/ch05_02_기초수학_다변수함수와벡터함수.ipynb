### 일변수 스칼라 함수
---

하나의 인풋(=스칼라, 일변수), 하나의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
y = f(x)
\end{align}

#### **예제 01**

\begin{align}
f(x) = x^2+2x+1
\end{align}

* 함수 생성하기

````python
def f1(x): # x: 스칼라
    return x**2 + 2*x + 1

f1(-1)
````

* 인풋 정의 (-10~10, 100개)

````python
# 100개 스칼라 값
xs = np.linspace(-10, 10, 100)
# 100개 스칼라 값에 대한 100개 스칼라 아웃풋
fs = f1(xs)
xs.shape, fs.shape
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(xs, fs)

# 설명 
ax.set_title('f(x) according to x')
ax.set_xlabel('x') # x축 라벨
ax.set_ylabel('f(x)') # y축 라벨

x = 5
ax.plot(x, f1(x), marker = 'o', c = 'r') # 특정 x에 따른 f(x) 표시
ax.text(x + 0.8, f1(x), f'({x}, { np.round(f1(x),2) })')
````

* 그래프 설명
  * x가 가질수 있는 값 : 모든 실수
  * f(x)가 가질 수 있는 값: 0 이상의 모든 실수
  * 그래프의 특징 : x가 -1일 때 아웃풋이 가장 작고, x가 -1에서 멀어질 때 아웃풋이 커진다.

#### **예제 02**

\begin{align}
f(x) = cosin(x)
\end{align}

* 함수 생성하기

````python
def f2(x):
    return np.cos(x)
````

* 인풋 정의 (-10~10, 100개)

````python
xs = np.linspace(-10,10,100)
````

* 2d플롯으로 시각화

````python
fig, ax = plt.subplots(1)
ax.plot(xs, f2(xs))

ax.set_title('$f(x)$ according to $x$', pad = 15)
ax.set_xlabel('$x$')
ax.set_ylabel('$f(x)$')

x = 0
ax.plot(x, f2(x), marker = 'o', color = 'navy')
ax.text(x + 0.5, f2(x), f'({x},{f2(x)})')

plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : -1 에서 1 사이 모든 실수 
  * 그래프의 특징 : x의 변화에 따라 아웃풋이 -1에서 1사이를 왔다갔다한다(횡보한다). 

### 다변수 스칼라 함수
---

여러개의 인풋(=다변수), 하나의 아웃풋(=스칼라, 스칼라함수)


\begin{align}
y = f(𝐱)\\
\end{align}



#### **예제 03**

\begin{align}
f(x, y) = x+y
\end{align}

* 함수 생성하기

````python
def f3(x, y): 
    return x + y

f3(2,2)
````

* 인풋 정의 (x, y, 모두 -15~20 값)

````python
xs = np.linspace(-15, 20, 36)
ys = np.linspace(-15, 20, 36)
print(xs.shape, ys.shape)

X, Y = np.meshgrid(xs, ys) # xs, ys의 순서쌍 모두 반환
print(X.shape, Y.shape)
F = f3(X, Y) # X, Y 모든 쌍에 대한 함수 결과 반환 
print(F.shape)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize = (9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X, Y, f3(X, Y), marker = '.', color = 'k') # x, y, z 축 데이터를 입력, 3차원 점 그리기
surf = ax.plot_surface(X, Y, f3(X,Y), cmap = 'rainbow', alpha = 0.7) # 점을 이은 면 surface 그리기
# 설명 
ax.set_title('f(x, y) according to x, y')
ax.set_xlabel('x') # x축 라벨
ax.set_ylabel('y') # y축 라벨
ax.set_zlabel('f(x,y)') # z축 라벨
# 컬러바
fig.colorbar(surf, label = 'f(x,y)', shrink = 0.6)
plt.show()
# x = 5
# ax.scatter3D(x, f1(x), marker = 'o', c = 'r') # 특정 x에 따른 f(x) 표시
# ax.text(x, f1(x), f'({x}, { np.round(f1(x),2) })')
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수
  * f(x)가 가질 수 있는 값 (아웃풋) : 모든 실수
  * 그래프의 특징 : x, y 값이 함께 증가하면 아웃풋도 커진다

#### **예제 04**

\begin{align}
f(x, y) = (x-5)^2+(y-5)^2 
\end{align}

* 함수 생성하기

````python
def f4(x, y):
    return (x - 5)**2 + (y - 5)**2
````

* 인풋 정의(x, y 모두 -15~20 사이 값)

````python
xs = np.linspace(-15, 20, 36)
ys = np.linspace(-15, 20, 36)

X, Y = np.meshgrid(xs, ys)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize = (9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X, Y, f4(X, Y), marker = '.', color = 'k') # x, y, z 축 데이터를 입력, 3차원 점 그리기
surf = ax.plot_surface(X, Y, f4(X,Y), cmap = 'rainbow', alpha = 0.7) # 점을 이은 면 surface 그리기
# 설명 
ax.set_title('f(x, y) according to x, y')
ax.set_xlabel('$x$') # x축 라벨
ax.set_ylabel('$y$') # y축 라벨
ax.set_zlabel('$f(x,y)$') # z축 라벨
# 컬러바
fig.colorbar(surf, label = 'f(x,y)', shrink = 0.6)
plt.show()
# x = 5
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수 
  * f(x)가 가질 수 있는 값 (아웃풋) : 0 이상의 모든 실수 
  * 그래프의 특징 : x, y == 5 일 때 최소 아웃풋을 가지고, x랑 y가 5에서 멀어질 수록 아웃풋이 커진다.

#### **예제 05**

\begin{align}
f(x_1, x_2) = 50(x_2-x_1^2)^2+(2-x_1)^2
\end{align}

* 함수 생성하기

````python
def f5(x1, x2):
    return 50 * (x2 - x1**2)**2 + (2 - x1)**2
````

* 인풋 정의 (x: -2 ~ 2 20개, y: -1~3 20개)

````python
xs = np.linspace(-2, 2, 20)
ys = np.linspace(-1, 3, 20)

X, Y = np.meshgrid(xs, ys)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize = (9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X, Y, f5(X, Y), marker = '.', color = 'k') # x, y, z 축 데이터를 입력, 3차원 점 그리기
surf = ax.plot_surface(X, Y, f5(X,Y), cmap = 'rainbow', alpha = 0.7) # 점을 이은 면 surface 그리기
# 설명 
ax.set_title('f(x, y) according to x, y')
ax.set_xlabel('x') # x축 라벨
ax.set_ylabel('y') # y축 라벨
ax.set_zlabel('f(x,y)') # z축 라벨
# 컬러바
fig.colorbar(surf, label = 'f(x,y)', shrink = 0.6)
plt.show()
````

* $x_1$, $x_2$의 범위를 둘다 -4부터 4로 변경하여 다시 그려보세요.

````python
xs = np.linspace(-4, 4, 20)
ys = np.linspace(-4, 4, 20)

X, Y = np.meshgrid(xs, ys)

# 3차원 플롯
fig = plt.figure(figsize = (9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(X, Y, f5(X, Y), marker = '.', color = 'k') # x, y, z 축 데이터를 입력, 3차원 점 그리기
surf = ax.plot_surface(X, Y, f5(X,Y), cmap = 'rainbow', alpha = 0.7) # 점을 이은 면 surface 그리기
# 설명 
ax.set_title('$f(x, y)$ according to $x$, $y$')
ax.set_xlabel('$x$') # x축 라벨
ax.set_ylabel('$y$') # y축 라벨
ax.set_zlabel('$f(x,y)$') # z축 라벨
# 컬러바
fig.colorbar(surf, label = '$f(x,y)$', shrink = 0.6)
plt.show()
````

* 그래프 설명
  * x, y가 가질수 있는 값 (인풋) : 모든 실수 
  * f(x)가 가질 수 있는 값 (아웃풋) : 0 이상의 모든 실수
  * 그래프의 특징 : x1 == 2, x2 ==4 일 때 최소 아웃풋 0을 가진다 (x1의 제곱 함수이기 때문에, x2의 변화보다 x1의 변화가 아웃풋 값에 더 큰 영향을 미친다). 

## 일변수 벡터함수
---

하나의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)




\begin{align}
f(x) = (f_1(x), f_2(x), ..., f_n(x))^T
\end{align}

#### **예제 06**

\begin{align}
f(x) = \begin{pmatrix}
h(x)\\ 
g(x)\\ 
\end{pmatrix}  = \begin{pmatrix}
x^2\\ 
5x\\ 
\end{pmatrix}
\end{align}

* 함수 $f(x)$ 생성하기

````python
def f6(x):
    return np.array([x**2,  5 * x])
````

* 인풋 정의 (-10~10 20개)

````python
xs = np.linspace(-10, 10, 20)  # x값 20개에 따른 
hs, gs = f6(xs)
fs = f6(xs)
xs.shape, hs.shape, gs.shape, fs.shape
````

* 2d플롯으로 시각화

````python
# output(f(x)) space 그려보기 
# x축을 h(x), y축을 g(x)

fig, ax = plt.subplots(1)
ax.plot(gs, hs) 

# 설명 
ax.set_title('f(x) space')
ax.set_xlabel('g(x)') # x축 라벨
ax.set_ylabel('h(x)') # y축 라벨

scat = ax.scatter(gs, hs, c = xs, cmap = 'rainbow')
fig.colorbar(scat)

plt.show()
````

* 그래프 설명
  * x가 가질수 있는 값 (인풋) : 모든실수
  * g(x)가 가질 수 있는 값 (첫번째 아웃풋) : 모든실수 
  * h(x)가 가질수 있는 값 (두번째 아웃풋): 0보다 큰 모든 실수 

#### **예제 07**

\begin{align}
f(t) = \begin{pmatrix}
x(t)\\ 
y(t)\\ 
z(t)\\ \end{pmatrix} = \begin{pmatrix}
sin(6t)\\
\frac{t}{4}\\
\frac{t^2}{2}\\
\end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f7(t):
    return np.array([np.sin(6*t), t/4, (t**2)/2])
````

* 인풋 정의 (-1~3 200개)

````python
xs =  np.linspace(-1, 3, 200)
xs.shape
````

* 3d플롯으로 시각화

````python
fig = plt.figure(figsize = (9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(f7(xs)[0], f7(xs)[1], f7(xs)[2], marker = '.', cmap = 'rainbow', c = xs, s = 40) # x, y, z 축 데이터를 입력, 3차원 점 그리기

ax.set_title('$f(t)$ space', pad = 15)
ax.set_xlabel('$x(t)$') # x축 라벨
ax.set_ylabel('$y(t)$') # y축 라벨
ax.set_zlabel('$z(t)$') # z축 라벨

fig.colorbar(surf, label = 't', shrink = 0.6)
plt.show()
````

* 그래프 설명
  * t가 가질수 있는 값 (인풋) : 모든 실수
  * x(t)가 가질 수 있는 값 (첫번째 아웃풋) : -1과 1 사이의 모든 실수
  * y(t)가 가질 수 있는 값 (두번째 아웃풋) : 모든 실수
  * z(t)가 가질 수 있는 값 (세번째 아웃풋) : 0 이상의 모든 실수 

#### **예제 08**

\begin{align}
\tau(t) = \begin{pmatrix}

z(t)\\ \end{pmatrix} = \begin{pmatrix}

  * sin(t)가 가질 수 있는 값 (두번째 아웃풋) : -1에서 1사이 모든 실수


## 다변수 벡터함수
---
여러개의 인풋(=스칼라, 일변수), 여러개의 아웃풋(=스칼라, 스칼라함수)

\begin{align}
f(𝐱) = (f_1(𝐱), f_2(𝐱), ..., f_n(𝐱))^T
\end{align}

#### **예제 09**


\begin{align}
f(x,y) = \begin{pmatrix}
x^2\\ 
y^2\\ 
xy\\ \end{pmatrix}
\end{align}

* 함수 생성하기

````python
def f9(x, y):
    return np.array([x**2, y**2, x * y])

f9(1,2)
````

* 인풋 정의 (x,y 모두 -0.5 ~ 1 사이 50개)

````python
xs = np.linspace(-0.5, 1, 50)
ys = np.linspace(-0.5, 1, 50)

# x, y  격자 순서쌍 
X, Y = np.meshgrid(xs, ys)
# 아웃풋 정의
A, B, C = f9(X,Y)
# 인풋 두개, 아웃풋 세개
X.shape, Y.shape, A.shape, B.shape, C.shape
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize = (9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(A, B, C, marker = '.', color = 'k') # x, y, z 축 데이터를 입력, 3차원 점 그리기
surf = ax.plot_surface(A, B, C, alpha = 0.5) # 점을 이은 면 surface 그리기
# 설명 
ax.set_title('f(x, y) space')
ax.set_xlabel('$x^2$') # x축 라벨
ax.set_ylabel('$y^2$') # y축 라벨
ax.set_zlabel('$xy$') # z축 라벨

plt.show()
````

#### **예제 10**


\begin{align}
f(u,v) = \begin{pmatrix}
u\\ 
v\\ 
1+u^2+\frac{v}{1+v^2}\\ \end{pmatrix}, \ \ \ \ \ 0 \leq u, v \leq1
\end{align}

* 함수 생성하기

````python
def f10(u,v):
    return np.array([u, v, 1 + u**2 + (v/(1+v**2))])
````

* 인풋 정의 (u, v 모두 0 ~ 1 사이 50개)

````python
us = np.linspace(0, 1, 50)
vs = np.linspace(0, 1, 50)

U , V = np.meshgrid(us, vs)
````

* 3d플롯으로 시각화

````python
# 3차원 플롯
fig = plt.figure(figsize = (9,7))
ax = plt.axes(projection = '3d') # 3차원 플롯
ax.scatter3D(f10(U,V)[0], f10(U,V)[1], f10(U,V)[2], marker = '.', color = 'k') # x, y, z 축 데이터를 입력, 3차원 점 그리기
surf = ax.plot_surface(f10(U,V)[0], f10(U,V)[1], f10(U,V)[2], alpha = 0.5) # 점을 이은 면 surface 그리기
# 설명 
ax.set_title('f(u, v) space')
ax.set_xlabel('$u$') # x축 라벨
ax.set_ylabel('$v$') # y축 라벨
ax.set_zlabel('$1 + u^2 + v/(1+v^2)$') # z축 라벨

plt.show()
````
